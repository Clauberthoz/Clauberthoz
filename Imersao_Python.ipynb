**Importacao**

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_principal = pd.read_excel("/Imersão Python - Tabela de ações (1).xlsx",sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/Imersão Python - Tabela de ações (1).xlsx",sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/Imersão Python - Tabela de ações (1).xlsx",sheet_name="Ticker")
df_ticker
```

```python
df_chat_gpt = pd.read_excel("/Imersão Python - Tabela de ações (1).xlsx",sheet_name="Chat gpt")
df_chat_gpt
```

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
# prompt: gere uma coluna com a variacao com o valor da porcentagem da coluna var_dia_pct

df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on= 'Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
# prompt: gere uma coluna no df_principal com o caluculo da variacao em reais

df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
# prompt: =IF(O3>0,"subiu",IF(O2<0,"desceu","estavel"))
# Tenho uma planilha de excel que usa essa formula, e eu gostaria de aplicar a mesma operacao em uma planilha do python. como ficaria o resultado em 1 linha?

df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: "subiu" if x > 0 else ("desceu" if x < 0 else "estavel"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on= 'Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt, left_on= 'Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (em anos)':'idade'}).copy()
df_principal
```

```python
# Calculando os valores
maior = df_principal["variacao_rs"].max()
menor = df_principal["variacao_rs"].min()
media = df_principal["variacao_rs"].mean()
media_subiu = df_principal[df_principal["resultado"] == "subiu"]["variacao_rs"].mean()
media_desceu = df_principal[df_principal["resultado"] == "desceu"]["variacao_rs"].mean()

# Exibindo os resultados
resultados = {
    "Maior": maior,
    "Menor": menor,
    "Média Geral": media,
    "Média de quem subiu": media_subiu,
    "Média de quem desceu": media_desceu
}

for chave, valor in resultados.items():
    print(f"{chave}: R$ {valor:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal["resultado"] == "subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("resultado")["variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
def formatar_numero(valor):
    if abs(valor) >= 1e9:
        return f'{valor/1e9:.2f}B'  # Bilhões
    elif abs(valor) >= 1e6:
        return f'{valor/1e6:.2f}M'  # Milhões
    elif abs(valor) >= 1e3:
        return f'{valor/1e3:.2f}K'  # Milhares
    return f'{valor:.2f}'  # Sem formatação

# Applying the formatting to the correct DataFrame (df_analise_saldo)
df_analise_saldo["variacao_formatada"] = df_analise_saldo["variacao_rs"].apply(formatar_numero)

fig = px.bar(df_analise_saldo,
             x="resultado",
             y="variacao_rs",
             text="variacao_formatada",  # Now using the formatted column from the correct DataFrame
             title="Variação em reais por resultado")

fig.update_traces(textposition='outside')  # Para colocar os valores fora das barras
fig.show()
```

```python

#df_analise_segmento_filt = df_analise_segmento[~df_analise_segmento['Segmento'].isin(["Energia e Petróleo", "Bancário", "Mineração"])]

fig = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Distribuição da Variação Financeira por Segmento')
fig.show()
```
